```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def check_url(url):
    try:
        # Send a request to the URL
        response = requests.get(url)

        # Check if the response is successful (HTTP status code within 200-299)
        if response.status_code >= 200 and response.status_code < 300:
            print(f"The URL {url} is working.")

            # If the response is successful, parse the page and find hyperlinks (backlinks)
            soup = BeautifulSoup(response.content, 'html.parser')
            links = soup.find_all('a', href=True)

            print("Backlinks on the page:")
            for link in links:
                # Construct absolute URLs for the backlinks
                backlink = urljoin(url, link['href'])
                print(backlink)

        else:
            print(f"The URL {url} returned an error: {response.status_code}")

    except requests.exceptions.RequestException as e:
        print(f"An error occurred while checking the URL {url}: {e}")

def check_multiple_urls(urls):
    for url in urls:
        check_url(url)
        print()

# URLs to check
urls_to_check = [
    "https://www.w3schools.com/sql/",
    "https://yournewsreporter.com/everything-you-should-know-about-hifu-treatment/",
    "https://thepereznotes.com/wool-and-the-circular-economy-a-symbiosis/",
]

# Call the function to check the URLs
check_multiple_urls(urls_to_check)
```
